## Need Cleaning

- mileage and km/kg done
- engine with nan and cc sub done
- max_power with bhp sub done
- seats has nan done
- torque needs removeing rpm and remove subs (Nm@,rpm,kgm,) done

```python
#df.drop(labels='name',axis=1,inplace=True)
df['name'] = df['name'].str.split(' ').str.slice(0,2).str.join(' ')
df['company'] = df['name'].str.split(' ').str.get(0)
```

```python
df['age'] = 2022-df['year']
df.drop(labels='year',axis=1,inplace=True)
```

```python
df['seats'].unique()
```

```python
df['mileage']=df['mileage'].str.split(' ').str.get(0)
```

```python
df['engine']=df['engine'].str.split(' ').str.get(0)
df=df[~df['engine'].isna()]
```

```python
df['max_power']= df['max_power'].str.split(' ').str.get(0)
```

```python
df=df[~df['torque'].isna()]
df['torque']=df['torque'].str.split('Nm',).str.get(0)
df['torque']=df['torque'].str.split('kgm').str.get(0)
df['torque']=df['torque'].str.split('@').str.get(0)
df['torque']=df['torque'].str.split('nm').str.get(0)
df['torque']=df['torque'].str.split('KGM').str.get(0)
df['torque']=df['torque'].str.split('NM').str.get(0)
```

```python
df.info()
```

```python
df.drop(df.index[df['torque']=="110(11.2)"],inplace=True)
df.drop(df.index[df['torque']=="210 / 1900"],inplace=True)

df['torque'].unique()
```

```python
df['mileage']=df['mileage'].astype(float)
df['engine']=df['engine'].astype(int)
df['max_power']=df['max_power'].astype(float)
df['torque']=df['torque'].astype(float)
df['seats']=df['seats'].astype(int)
```

```python
df.drop(labels='name',axis=1,inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df=df[df['selling_price']<9e6].reset_index(drop=True)
```

## Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df.info()
```

```python
cat_df = df.select_dtypes(include=['object'])
print(cat_df)
dummies = pd.get_dummies(cat_df,drop_first=True)
df = df.drop(list(cat_df.columns),axis=1)
df = pd.concat([df,dummies],axis=1)
```

```python
scalar = MinMaxScaler()
df = pd.DataFrame(scalar.fit_transform(df),columns=df.columns,index=df.index)
df.head()
```

```python
X = df.drop(columns='selling_price')
y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_train_pred = lr.predict(X_train)
```

```python
r2_train_pred = r2_score(y_train_pred,y_train)
print(r2_train_pred)
```

```python
y_test_pre = lr.predict(X_test)
```

```python
r2_test_pred = r2_score(y_test_pre,y_test)
print(r2_test_pred)
```
