```python
#imports 
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('car data.csv')
df.shape
```

```python
df['age'] = 2022-df['Year']
df.drop(['Car_Name'],axis=1,inplace=True)
df.head()
```

```python
df.drop(['Year'],axis=1,inplace=True)
```

```python
categorical_df = df.select_dtypes(include=['object'])
dummies = pd.get_dummies(categorical_df,drop_first=True)
df = df.drop(list(categorical_df.columns),axis=1)
df = pd.concat([df,dummies],axis=1)
df
```

```python
scalar = MinMaxScaler()
df = pd.DataFrame(scalar.fit_transform(df),columns=df.columns,index=df.index)
df.head()
```

```python
X = df
y = df.pop('Selling_Price')
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_train_pred = reg.predict(X_train)
print(y_train_pred)
print(y_train)
```

```python
r2_score_pred = r2_score(y_train_pred,y_train)
print(r2_score_pred)
```

```python
y_test_pred = reg.predict(X_test)
print(y_test_pred)
print(y_test)
r2_testscore_pred = r2_score(y_test,y_test_pred)
print(r2_testscore_pred)
```
